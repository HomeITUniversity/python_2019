# Задача 1 (парсим дату)

Вы ежденевно получаете отчет, в котором разработчики отчета вставили дату и время в удобном для них формате.
При считывании из файла, ваша программа читает строчку из файла в формте str - строка.

Пример даты из отчета - "D: 2019-21-10 T: 21_55"
Вам необходимо реализовать функцию, которая будет преобразовывать прочитанную из файла дату в приемлимую для вас дату.
Приемлимая дата - "21.10.2019 21:55"

#### Решение

```python
from datetime import datetime

readed_string = "D: 2018-21-10 T: 21_55"
need_string = "21.10.2019 21:55:00"

def date_convertor(string_with_date):
    date = datetime.strptime(string_with_date, "D: %Y-%d-%m T: %H_%M")
    print(date.strftime("%d.%m.%Y %H:%M:%S"))

date_convertor(readed_string)
```

# Задача 2 (бесконечные циклы и работа по времени)

Вы пишите программу имитирующую управление самолетом.<br>
Руль самолета имеет возможность перемещаться вверх - вниз, влево - вправо.<br>
Прибор считывающий положение руля возвращает вам 2 координаты (-90 >= x >= 90) (-90 >= y >= 90).<br>
Соответственно 90 - поворот на 90 градусов вверх или вправо, -90 поворот на 90 градусов вниз или влево.<br>
На поворот в заданные координаты у самолета уходит |0.1y| + |0.2x| секунд. <br>

Если наклон самолета по любому из направлений составляет больше 75 градусов, то самолет уходит в штопор и вернуть его в нормальное состояние не становится возможным.<br>

Полет длится до первого падения, но не более 60 секунд.<br>
Если самолет пролетел 60 и более секунд без аварии, то вывести что тест сдан, иначе вывести что тест провален.<br>
Инертностью работы функций пренебречь.

Ваша задача написать 3 функции.<br>

1) Эмулятор поворота самолета.<br>
2) Функцию полета самолета. Функция должна выдавать предупреждение о возможном сваливании и запрашивать продолжение такого поврота путем ввода с клавиатуры "Y" для подтверждения.<br>
3) Функцию выводящую результат полета самолета с указанием времени полета, количество удачных поворотов, сдал/не сдал<br>



#### Решение

```python
import math, time, random
from datetime import datetime, timedelta
def skyship_controller(x, y):
    t = math.ceil(math.fabs(y) * .1 + .2 * math.fabs(x))
    
    print("начинаю поворот {coordinate}\nвремя на поворот: {time} секунд".format(
        time=t,
        coordinate="{updown} и {rightleft}".format(
            updown="прямо" if y == 0 else "вниз " if y < 0 else "вверх",
            rightleft="прямо" if x == 0 else "влево" if x < 0 else "вправо" 
        )
    ))
    
    time.sleep(t)
    print("завершил поворот")
    return t

def signal_reader():
    return random.randint(-90, 90), random.randint(-90, 90)

def fly():
    
    is_pass = False
    correct = 0
    flight_time = 0
    
    while True:
        if flight_time >= 60:
            is_pass = True
            break
        print("летим")
        
        x, y = signal_reader()
        
        if math.fabs(x) > 75 or math.fabs(y) > 75:
            print("предупреждение, поворот на угол x:{x} y:{y} приведет к сваливанию.".format(x=x, y=y))
            access = input("для продолжения напишите Y, для отмены напишите любую другую букву")
            if access == "Y":
                flight_time += skyship_controller(x,y)
                print("Самолет ушел в штопор")
                is_pass = False
                break
        else:
            flight_time += skyship_controller(x,y)
            correct += 1
            
        t = random.randint(1,10)
        print("летим прямо в течении {} секунд".format(t))    
        time.sleep(t)
    
    return is_pass, correct, flight_time

def analysis(fly):
    is_pass, correct, flight_time = fly
    print("Сдан" if is_pass else "Не сдан", f"Верных поворотов: {correct}", f"Время полета {flight_time} секунд")
    
        

analysis(fly())
```
