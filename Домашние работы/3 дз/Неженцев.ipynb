Поэтому: не теряй времени, выбирай слова, не упускай возможность. – Конфуций.» Необходимо получить список(без знаков препинания) отдельных слов и вывести его (список). Затем необходимо вывести все слова в обратном порядке, не изменяя список.

```python
#напишите ваше решение сюда
a = "Три вещи никогда не возвращаются обратно — время, слово, возможность. Поэтому: не теряй времени, выбирай слова, не упускай возможность. – Конфуций."
symbol = ['.', ',', '– ', '"', ':', '—']
for i in symbol:
  a = a.replace(i, "")
  b = list(a.split())
print (b)
b.reverse()
print (b)
#Надеюсь правильно понял и разворачивать надо список, а не строку до форматирования
```
