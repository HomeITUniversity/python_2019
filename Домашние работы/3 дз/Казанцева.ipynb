# Задача 2

Поэтому: не теряй времени, выбирай слова, не упускай возможность. – Конфуций.» Необходимо получить список(без знаков препинания) отдельных слов и вывести его (список). Затем необходимо вывести все слова в обратном порядке, не изменяя список.

```python
#напишите ваше решение сюда
s = 'Три вещи никогда не возвращаются обратно — время, слово, возможность. \
Поэтому: не теряй времени, выбирай слова, не упускай возможность. – Конфуций.'
s1 = ''
for_del = ['.', ',', ';', ':', '— ', '– ', '!', '?']
for i in for_del:
  s = s.replace(i, '')
s1 = s.split(' ')
print(s1)
print(s1[::-1])
```

# Задача 4
Найти максимальное  число из списка, которое делится на 7 без остатка. [random list]. Если такого нет, то вывести сообщение об этом.

```python
#напишите ваше решение сюда
import random
random_list = [random.randint(0, 100) for i in range(10)]
print(random_list)

list7 = []
for i in range(len(random_list)):
  if random_list[i] % 7 == 0:
    list7.append(random_list[i])
print(list7)
if list7 != []:
  print('Максимальное из делящихся на 7 без остатка:', max(list7))
else:
  print('В списке нет чисел, которые делятся на 7 без остатка')
```

# Задача 5
Дан лист кортежей L=[(«Иван», 24), («Марина», 44), («Виолетта», 6), («Александр», 30), («Виталик», 25)] Необходимо отсортировать список по возрасту.

```python
#напишите ваше решение сюда
L=[("Иван", 24), ("Марина", 44), ("Виолетта", 6), ("Александр", 30), ("Виталик", 25)]
print(sorted(L, key=lambda age: age[1]))

#И вариант решения, если не нужно сохранять первоначальное состояние списка: 
#L.sort(key=lambda age: age[1])
#print(L)
```

```python
#Это еще один вариант решения на случай, если нельзя было использовать lambda
L=[("Иван", 24), ("Марина", 44), ("Виолетта", 6), ("Александр", 30), ("Виталик", 25)]
L2 = []
for i in range(len(L)):
  L2.append(L[i][1])
L2.sort()
for i in range(len(L2)):
  for j in range(len(L)):
    if L[j][1] == L2[i]:
      L2[i] = L[j]
print(L2)

  
```
