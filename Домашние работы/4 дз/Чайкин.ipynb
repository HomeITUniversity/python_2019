# Задача 4

Маша и Аня работают в отделе продаж нашего банка и продают Кредиты наличными. <br>
Оформленные ими кредиты представлены в соответствующих словарях (сумма кредита, ставка, срок в годах)<br>
Среди клиентов Маши и Ани найти клиентов и вывести на экран, у которых:<br>
•	Совпадает ставка<br>
•	Совпадает сумма кредита<br>
•	Совпадает срок выплаты<br>
•	Не совпадает сумма, но совпадает срок выплаты<br>
•	Совпадает процентная ставка, но не совпадает сумма кредита.<br>
Условие может, как выполниться, так и не выполниться (такое пересечение может быть пустым)<br>


```python
masha_credits = {
        'Салтыков Иван': {
        'credit': 80000,
        'percent': 22.5,
        'period': 3
    },
        'Иванова Марина': {
        'credit': 240000,
        'percent': 20.8,
        'period': 5
    },
        'Жученко Виталий': {
        'credit': 35000,
        'percent': 19.4,
        'period': 2
    },
        'Петров Александр': {
        'credit': 100000,
        'percent': 27.9,
        'period': 3
    },
        'Краснов Антон': {
        'credit': 10000,
        'percent': 55.2,
        'period': 1
    },
        'Полено Елена': {
        'credit': 330000,
        'percent': 4.0,
        'period': 1
    },
}

ann_credits = {
        'Букин Геннадий': {
        'credit': 1000,
        'percent': 55.2,
        'period': 20
    },
        'Антонов Юрий': {
        'credit': 20000,
        'percent': 20.8,
        'period': 4
    },
        'Трененков Виталий': {
        'credit': 35000,
        'percent': 20.1,
        'period': 1
    },
        'Самоваров Павел': {
        'credit': 100000,
        'percent': 27.9,
        'period': 2
    },
        'Самсонов Владимир': {
        'credit': 76000,
        'percent': 11.9,
        'period': 4
    },
        'Антонина Антонина': {
        'credit': 1100000,
        'percent': 4.0,
        'period': 5
    },
        'Колованов Ярослав': {
        'credit': 25000,
        'percent': 22.2,
        'period': 2
    },
        'Чайкина Наталья': {
        'credit': 330000,
        'percent': 18.9,
        'period': 3
    },
}

# ваше решение тут

#Задаю словари что бы читаемо вывести, у кого что совпадает, а что нет

percent={}
credit={}
period={}
nocredit_period={}
percent_nocredit={}


#Сначала ищу пересечения в разных словарях

for key, val in masha_credits.items():
    for key2, val2 in ann_credits.items():
        # Если совпадает ставка
        if val['percent'] == val2['percent']:
            percent[val['percent']] = [key,key2]

        # Если совпадает сумма
        if val['credit'] == val2['credit']:
            credit[val['credit']]= [key,key2]

        # Если совпадает срок
        if val['period'] == val2['period']:
            period[val['period']] = [key,key2]
            
        # Если совпадает срок но не совпадет сумма
        if val['period'] == val2['period'] and val['credit'] != val2['credit']:
            nocredit_period[val['period']]=[key,key2]   
            
        # Если совпадает ставка но не совпадет сумма   
        if val['percent'] == val2['percent'] and val['credit'] != val2['credit']:
            percent_nocredit[val['percent']] = [key,key2]    

#Потом в первом     
            
for key, val in masha_credits.items():
    for key2, val2 in masha_credits.items():
        # Если совпадает ставка
        if key!=key2 and val['percent'] == val2['percent']:
            if percent[val['percent']].count(key)==0:
                percent[val['percent']].append(key)
            elif percent[val['percent']].count(key2)==0:
                percent[val['percent']].append(key2)

        # Если совпадает сумма
        if key!=key2 and val['credit'] == val2['credit']:
            if credit[val['credit']].count(key)==0:
                credit[val['credit']].append(key)
            elif credit[val['credit']].count(key2)==0:
                credit[val['credit']].append(key2)

        # Если совпадает срок
        if key!=key2 and val['period'] == val2['period']:
            if period[val['period']].count(key)==0:
                period[val['period']].append(key)
            elif period[val['period']].count(key2)==0:
                period[val['period']].append(key2)
                 
        # Если совпадает срок но не совпадет сумма 
        if key!=key2 and val['period'] == val2['period'] and val['credit'] != val2['credit']:
            if nocredit_period[val['period']].count(key)==0:
                nocredit_period[val['period']].append(key)
            elif nocredit_period[val['period']].count(key2)==0:
                nocredit_period[val['period']].append(key2)  
            
        # Если совпадает ставка но не совпадет сумма   
        if key!=key2 and val['percent'] == val2['percent'] and val['credit'] != val2['credit']:
            if percent_nocredit[val['period']].count(key)==0:
                percent_nocredit[val['period']].append(key)
            elif percent_nocredit[val['period']].count(key2)==0:
                percent_nocredit[val['period']].append(key2) 
                
#Потом во втором             
            
for key, val in ann_credits.items():
    for key2, val2 in ann_credits.items():
        # Если совпадает ставка
        if key!=key2 and val['percent'] == val2['percent']:
            percent.setdefault(val['period'],[key])
            if percent[val['percent']].count(key)==0:
                percent[val['percent']].append(key)
            elif percent[val['percent']].count(key2)==0:
                percent[val['percent']].append(key2)

        # Если совпадает сумма
        if key!=key2 and val['credit'] == val2['credit']:
            credit.setdefault(val['period'],[key])
            if credit[val['credit']].count(key)==0:
                credit[val['credit']].append(key)
            elif credit[val['credit']].count(key2)==0:
                credit[val['credit']].append(key2)

        # Если совпадает срок
        if key!=key2 and val['period'] == val2['period']:
            period.setdefault(val['period'],[key])
            if period[val['period']].count(key)==0:
                period[val['period']].append(key)
            elif period[val['period']].count(key2)==0:
                period[val['period']].append(key2)
                 
        # Если совпадает срок но не совпадет сумма 
        if key!=key2 and val['period'] == val2['period'] and val['credit'] != val2['credit']:
            nocredit_period.setdefault(val['period'],[key])
            if nocredit_period[val['period']].count(key)==0:
                nocredit_period[val['period']].append(key)
            elif nocredit_period[val['period']].count(key2)==0:
                nocredit_period[val['period']].append(key2)  
            
        # Если совпадает ставка но не совпадет сумма   
        if key!=key2 and val['percent'] == val2['percent'] and val['credit'] != val2['credit']:
            percent_nocredit.setdefault(val['period'],[key])
            if percent_nocredit[val['period']].count(key)==0:
                percent_nocredit[val['period']].append(key)
            elif percent_nocredit[val['period']].count(key2)==0:
                percent_nocredit[val['period']].append(key2) 


#Итог          
                
for i,j in percent.items():
    print(f"одинаковая ставка {i} у {j}")

print("\n")
    
for i,j in credit.items():
    print(f"одинаковый кредит {i} у {j}")
    
print("\n")
    
for i,j in period.items():
    print(f"одинаковый срок {i} у {j}")
    
print("\n")
    
for i,j in nocredit_period.items():
    print(f"одинаковый срок {i}, но разный кредит у {j}")
    
print("\n")
    
for i,j in percent_nocredit.items():
    print(f"одинаковая ставка {i}, но разный кредит у {j}")
    
#print(f"{percent},\n{credit},\n{period},\n{nocredit_period},\n{percent_nocredit}") 
```

# Задача 5
Суть задачи в том, чтобы из англо-латинского словаря сделать латино-английский.

Входные данные:<br>
apple - malum, pomum, popula<br>
fruit - baca, bacca, popum<br>
punishment - malum, multa<br>

Выходные данные:<br>
baca - fruit<br>
bacca - fruit<br>
malum - apple, punishment<br>
multa - punishment<br>
pomum - apple<br>
popula - apple<br>
popum - fruit<br>

```python
# ваше решение тут



#Задаем англо латинский словарь

anglo_latine = {
        'apple': ['malum', 'pomum', 'popula'],
        'fruit': ['baca', 'bacca', 'popum'],
        'punishment': ['malum', 'multa']
               }
 
  
#Делаем латино-английский
latine_anglo={}

  
for key,val in anglo_latine.items():
    for i in val:
        if latine_anglo.get(i)==None:
            latine_anglo[i]=[key]
        else:
            latine_anglo[i].append(key)

            
for key,val in latine_anglo.items():     
    print(key,val)
print(latine_anglo)
 
```
