# Задача 1

Нужно составить словарь, в котором ключом будет ФИО, а значением возраст.

# Задача 5
Суть задачи в том, чтобы из англо-латинского словаря сделать латино-английский.

Входные данные:<br>
apple - malum, pomum, popula<br>
fruit - baca, bacca, popum<br>
punishment - malum, multa<br>

Выходные данные:<br>
baca - fruit<br>
bacca - fruit<br>
malum - apple, punishment<br>
multa - punishment<br>
pomum - apple<br>
popula - apple<br>
popum - fruit<br>

```python
# ваше решение тут
al = {"apple":["malum", "pomum", "popula"], "fruit":["baca", "bacca", "popum"], "punishment":["malum", "multa"]}
la = {}
lst=[]
r=[]
for k, v in al.items():
  lst.append(k)
  for i  in v:
    if not(i in la):
      la[i]=lst
    else:
      r=la[i][:]
      r.append(k)
      la[i]=r[:]
  lst=[]
for i,j in la.items():
  print(i, "-", ", ".join(j))
```
