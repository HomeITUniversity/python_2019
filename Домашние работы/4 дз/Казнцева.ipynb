# Задача 4

Маша и Аня работают в отделе продаж нашего банка и продают Кредиты наличными. <br>
Оформленные ими кредиты представлены в соответствующих словарях (сумма кредита, ставка, срок в годах)<br>
Среди клиентов Маши и Ани найти клиентов и вывести на экран, у которых:<br>
•	Совпадает ставка<br>
•	Совпадает сумма кредита<br>
•	Совпадает срок выплаты<br>
•	Не совпадает сумма, но совпадает срок выплаты<br>
•	Совпадает процентная ставка, но не совпадает сумма кредита.<br>
Условие может, как выполниться, так и не выполниться (такое пересечение может быть пустым)<br>


```python
masha_credits = {
        'Салтыков Иван': {
        'credit': 80000,
        'percent': 22.5,
        'period': 3
    },
        'Иванова Марина': {
        'credit': 240000,
        'percent': 20.8,
        'period': 5
    },
        'Жученко Виталий': {
        'credit': 35000,
        'percent': 19.4,
        'period': 2
    },
        'Петров Александр': {
        'credit': 100000,
        'percent': 27.9,
        'period': 3
    },
        'Краснов Антон': {
        'credit': 10000,
        'percent': 55.2,
        'period': 1
    },
        'Полено Елена': {
        'credit': 330000,
        'percent': 4.0,
        'period': 1
    },
}

ann_credits = {
        'Букин Геннадий': {
        'credit': 1000,
        'percent': 55.2,
        'period': 20
    },
        'Антонов Юрий': {
        'credit': 20000,
        'percent': 20.8,
        'period': 4
    },
        'Трененков Виталий': {
        'credit': 35000,
        'percent': 20.1,
        'period': 1
    },
        'Самоваров Павел': {
        'credit': 100000,
        'percent': 27.9,
        'period': 2
    },
        'Самсонов Владимир': {
        'credit': 76000,
        'percent': 11.9,
        'period': 4
    },
        'Антонина Антонина': {
        'credit': 1100000,
        'percent': 4.0,
        'period': 5
    },
        'Колованов Ярослав': {
        'credit': 25000,
        'percent': 22.2,
        'period': 2
    },
        'Чайкина Наталья': {
        'credit': 330000,
        'percent': 18.9,
        'period': 3
    },
}

# ваше решение тут

from copy import deepcopy
d = deepcopy(masha_credits)
d.update(ann_credits)

p = set()
c = set()
s = set()
c_s = set()
p_c = set()

for kl, val in d.items():
  for k, v in d.items():
    if v.get('percent') == val.get('percent') and k != kl:
      p.add(k)
    if v.get('credit') == val.get('credit') and k != kl:
      c.add(k)
    if v.get('period') == val.get('period') and k != kl:
      s.add(k)
    if v.get('credit') != val.get('credit') and v.get('period') == val.get('period') and k != kl:
      c_s.add(k)
    if v.get('percent') == val.get('percent') and v.get('credit') != val.get('credit') and k != kl:
      p_c.add(k)
print('Клиенты, у которых есть совпадения по ставке (неуникальная ставка):')
print(p, '\n')
print('Клиенты, у которых есть совпадения по сумме кредита (неуникальная сумма кредита):')
print(c, '\n')
print('Клиенты, у которых есть совпадения по сроку выплаты (неуникальный срок выплаты):')
print(s, '\n')
print('Клиенты, у которых нет совпадения по сумме кредита, но есть совпадения по сроку выплаты:')
print(c_s, '\n')
print('Клиенты, у которых есть совпадения по ставке, но нет совпадения по сумме кредита:')
print(p_c, '\n')
```

# Задача 5
Суть задачи в том, чтобы из англо-латинского словаря сделать латино-английский.

Входные данные:<br>
apple - malum, pomum, popula<br>
fruit - baca, bacca, popum<br>
punishment - malum, multa<br>

Выходные данные:<br>
baca - fruit<br>
bacca - fruit<br>
malum - apple, punishment<br>
multa - punishment<br>
pomum - apple<br>
popula - apple<br>
popum - fruit<br>

```python
# ваше решение тут

s0 = 'apple - malum, pomum, popula'.split(' - ') # Разбила строку на будущие ключ и значение
s1 = 'fruit - baca, bacca, popum'.split(' - ')
s2 = 'punishment - malum, multa'.split(' - ')
d = {}

if s0[1].find(', ') != -1:    # Если значение состоит из нескольких слов, разделенных запятой, то превращаем строку в список
  s0[1] = s0[1].split(', ')   # таким образом если значение - это одно слово, то оно и останется словом, а не превратится в список
if s1[1].find(', ') != -1:
  s1[1] = s1[1].split(', ')
if s2[1].find(', ') != -1:
  s2[1] = s2[1].split(', ')            

d = {                         # формируем входной словарь
    s0[0]: s0[1],
    s1[0]: s1[1],
    s2[0]: s2[1]
}

print(d)

d1 = {}                       # d1 - выходной словарь

for k, v in d.items():        # берем пары ключ/значение входного словаря
  if type(v) is list:         # если в качестве значения имеем список слов, то бежим по списку:
    for i in range(len(v)):
      if v[i] not in d1:                 # и прежде чем добавить ключ/значение в выходной словарь проверяем есть ли в нем уже такой ключ
        d1[v[i]] = k                     # если нет, то просто добавляем ключ/значение
      else:                              # если есть, то проверяем тип значения (список или строка)
        if type(d1[v[i]]) is str:        # если это строка, то превращаем её в список и добавляем еще одно значение
          l = []
          l.append(d1[v[i]])
          l.append(k)
          d1[v[i]] = l
        else:                            # если это список, то просто добавляем к нему еще одно значение
          d1[v[i]].append(k)
  else:                        # если в качестве значения у нас не список, а строка,
    if v not in d1:                  # то проверяем есть ли такой ключ в выходном списке и добавляем ключ/значение, если не нашли
      d1[v] = k
    else:                            # а если такой ключ нашли в выходном списке, то смотрим строка это или список, чтобы добавить еще одно значение
      if type(d1[v]) is str:
          l = []
          l.append(d1[v])
          l.append(k)
          d1[v] = l
      else:
        d1[v].append(k)

print(d1) 
```
