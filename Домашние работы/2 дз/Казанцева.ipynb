# Задача 1
Дана строка (вводится с клавиатуры).

Найдите наиболее часто встречаемый символ в строке. Отобразить в консоль сам символ и частоту его появления.

```python
# Ваше решение
s = str(input('Введите строку: '))
k = s.count(s[0])
ks = s[0]
for i, char in enumerate(s):
  if s.count(s[i]) > k:
    k = s.count(s[i])
    ks = s[i]
  elif s.count(s[i]) == k and s[i] not in ks:
    ks = ks + ', ' + s[i]
print('Наиболее часто встречаемый символ:', ks, '\nчастота появления:', k)
```

# Задача 2

```python
# Ваше решение
s = str(input('Введите пароль: '))
digit = 0
down = 0
up = 0
for i in range(len(s)):
  if s[i].isdigit() == True:
    digit = 1
  if s[i].islower() == True:
    down = 1
  if s[i].isupper() == True:
    up = 1
if len(s) > 10 and digit == 1 and down == 1 and up == 1:
  print('Пароль корректный')
else:
  print('''Проверьте корректность пароля!
Пароль должен содержать как минимум 1 цифру, 1 букву нижнего регистра
и 1 букву верхнего регистра и быть длинной > 10 символов.''')
```

# Задача 4
С клавиатуры вводится строка.

Необходимо удалить из строки все пробелы и определить, является ли она перевертышем (палиндромом).

**Пример палиндрома:** "Кот учён, но не чуток"

```python
# Ваше решение
s = str(input('Введите строку: ')).lower()
for_del = [' ', '.', ',', ';', ':', '-', '!', '?']
for i in for_del:
  s = s.replace(i, '')
s1 = s[::-1]
if s == s1:
  print('Палиндром')
else:
  print ('Не палиндром')
```

Необходимо удалить из строки повторяющиеся символы. Вывести отдельно строку без повторяющихся символов. Пробелы оставить на своих местах.
