# Задача 2
С клавиатуры вводится пароль.

Необходимо проверить корректность пароля. Пароль должен содержать как минимум 1 цифру, 1 букву нижнего регистра и 1 букву верхнего регистра и быть длинной > 10 символов. 

```python
# Ваше решениe
p = input ("Введите пароль: ")
while True:
    if p.isdigit():
      print ("не хватает букв")
    if p.isalpha():
      print("не хватает цифр")
    if p.islower():
      print("не хватает букв ВЕРХНЕГО регистра")
    if p.isupper():
      print("не хватает букв нижнего регистра")
    if len(p) < 10:
      print("пароль слишком короткий")
    else: print("пароль принят")
    break
```
